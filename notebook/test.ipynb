```python
def get_title_text(url, title):
    loader = UnstructuredLoader(web_url=url)
    data = loader.load()
    narrative_text = [el.page_content for el in data if el.metadata.get('category') == 'NarrativeText']
    return [title]+narrative_text
```

```python
model_name_or_path = "../multilingual-e5-large-instruct"
model = SentenceTransformer(model_name_or_path, trust_remote_code=True)
```

```python
def get_embeddings(content):

    # embeddings = ollama.embed(model='nomic-embed-text:latest',input=content)
    # embeddings = torch.tensor(embeddings.embeddings)
    # # text_embeddings = torch.tensor(embeddings.embeddings[1:])
    # embeddings = F.normalize(embeddings, p=2, dim=1)
    
    embeddings = model.encode(content, convert_to_tensor=True, normalize_embeddings=True,show_progress_bar=False)

    return embeddings
    
```

```python
def get_similarity_scores(embeddings):
    
    # embeddings = torch.tensor(embeddings.embeddings)
    # # text_embeddings = torch.tensor(embeddings.embeddings[1:])
    # embeddings = F.normalize(embeddings, p=2, dim=1)
    # title_embedding = F.normalize(title_embedding, p=2, dim=1)
    scores = (embeddings[:1] @ embeddings[1:].T) * 100

    return scores
```
